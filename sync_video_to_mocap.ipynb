### find the correspondence 

```python
def map_video_frame_to_mocap(
    video_frame_idx,
    videoFrame1, mocapFrame1,
    video_fps=30, mocap_fps=100
):
   
    #calculate the time by using the video frame 
    frame_diff = video_frame_idx - videoFrame1
    time_diff = frame_diff / video_fps

    #calculated the realted mocape index
    mocap_frame_diff = time_diff * mocap_fps
    mocap_frame = mocapFrame1 + mocap_frame_diff

    #round the index
    return int(round(mocap_frame))


rgb_first_frame = first_frame_index
rgb_last_frame = first_stop_frame_index
mocap_first_frame = mocap_first_frame

#design the correspondance dic only for moving points 
moving_correspondence = {}
for video_frame in range (rgb_first_frame, rgb_last_frame):
    curr_corr_mocap_frame_idnex = map_video_frame_to_mocap(video_frame, rgb_first_frame, mocap_first_frame)
    moving_correspondence[video_frame] = curr_corr_mocap_frame_idnex # index - index

correspondence_num = len(moving_correspondence.keys())
print("correspondence_num is: ", correspondence_num)

print(moving_correspondence)
```

### find the 2D points for each frame 

```python
import cv2
import apriltag
import csv
def detect_apriltag_center():
    frames = [cv2.imread(os.path.join(RGB_file_path,file)) for file in sorted(os.listdir(RGB_file_path))]
    # print(len(frames))
    detector = apriltag.Detector()
    apriltag_centers = {}
    for index, frame in enumerate(frames):
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        result = detector.detect(gray)
        if len(result) == 0:
            center_x, center_y = (-1,-1)
            apriltag_centers[index] = (center_x, center_y)
            print("!!!!!!!!")
        else:
            detection = result[0]
            center_x, center_y = detection.center
            apriltag_centers[index] = (center_x, center_y)
    return apriltag_centers


apriltag_centers = detect_apriltag_center()

print("the total apriltag center number is: ", len(apriltag_centers)) #2265
```

### find the 50 samples 

```python
import numpy as np
def evenly_sample_correspondences(PNP_dataset, num_samples):
   
    total = len(PNP_dataset)

    indices = np.round(np.linspace(0, total - 1, num_samples)).astype(int)

  

    sampled_dataset = [PNP_dataset[i] for i in indices]
    return sampled_dataset


PNP_50 = evenly_sample_correspondences(PNP_dataset, 50)

print(len(PNP_50))
print(PNP_50)
```

### run the pnp 

```python
points_2D = [i[0] for i in PNP_50]
points_3D = [i[1] for i in PNP_50]

#change the type
points_3D = np.array(points_3D, dtype=np.float32)
points_2D = np.array(points_2D, dtype=np.float32)

print("points_3D shape is: ", points_3D.shape)
print("points_2D shape is: ", points_2D.shape)



dist_coeffs = None
K = np.array([
    [617.552978515625, 0.0, 323.6305236816406],
    [0.0, 618.0050659179688, 242.8762664794922],
    [0, 0, 1]

], dtype = np.float32)



success, R_vec, t = cv2.solvePnP(
    points_3D,
    points_2D,
    K,
    dist_coeffs,
    cv2.SOLVEPNP_EPNP
)

# success, rvec, tvec, inliers = cv2.solvePnPRansac(
#     points_3D,       # (N,3) float32/float64
#     points_2D,       # (N,2) float32/float64
#     K,                
#     dist_coeffs,     
#     flags=cv2.SOLVEPNP_ITERATIVE,
#     reprojectionError=100.0,  # RANSAC inlier threshold 
#     confidence=0.99,        # confidence level
#     iterationsCount=100     # RANSAC iteration time 
# )


R_matrix, _ = cv2.Rodrigues(R_vec)

print("transportation is: ", t)
```
